```python
# 获取数据
import requests
import json

headers = {
    'server': 'wswaf'
}

response = requests.get('https://www.cwl.gov.cn/cwl_admin/front/cwlkj/search/kjxx/findDrawNotice?name=ssq&issueCount=&issueStart=&issueEnd=&dayStart=&dayEnd=&pageNo=1&pageSize=2000&week=&systemType=PC', headers=headers)

if response.status_code == 200:
    # print(response.json())
    print(response.json()['total'])
    print(len(response.json()['result']))

    json_str = json.dumps(response.json(), indent=4)
    with open("data/svm-ssq/data.json", "w", encoding="utf-8") as f:
        f.write(json_str)
```

```python
import json
def week_to_number(week):
    if week == '一':
        return 0
    if week == '二':
        return 1
    if week == '三':
        return 2
    if week == '四':
        return 3
    if week == '五':
        return 4
    if week == '六':
        return 6
    if week == '七':
        return 6
    return -1
    
    
with open("data/svm-ssq/data.json", "r", encoding="utf-8") as f:
    data = json.load(f)
    
for info in data['result']:
    print(info['date'].replace('(', ' ').replace(')', '').split(' ')[0].split('-'))
    print(week_to_number(info['week']))
    print(info['red'].split(','))
    print(info['blue'])
    print(info['content'].split(','))
    print()
```

```python
import sxtwl
def outTianGanDiZhi(year, month, day):
    day = sxtwl.fromSolar(year, month, day)
    Gan = ["甲", "乙", "丙", "丁", "戊", "己", "庚", "辛", "壬", "癸"]
    Zhi = ["子", "丑", "寅", "卯", "辰", "巳", "午", "未", "申", "酉", "戌", "亥"]
    tianGanDiZhi_years = ['甲子', '乙丑', '丙寅', '丁卯', '戊辰', '己巳', '庚午', '辛未', '壬申', '癸酉',
                          '甲戌', '乙亥', '丙子', '丁丑', '戊寅', '己卯', '庚辰', '辛巳', '壬午', '癸未',
                          '甲申', '乙酉', '丙戌', '丁亥', '戊子', '己丑', '庚寅', '辛卯', '壬辰', '癸巳',
                          '甲午', '乙未', '丙申', '丁酉', '戊戌', '己亥', '庚子', '辛丑', '壬寅', '癸卯',
                          '甲辰', '乙巳', '丙午', '丁未', '戊申', '己酉', '庚戌', '辛亥', '壬子', '癸丑',
                          '甲寅', '乙卯', '丙辰', '丁巳', '戊午', '己未', '庚申', '辛酉', '壬戌', '癸亥']

    def getGZIndex(tgDz):
        for i in range(len(tianGanDiZhi_years)):
            if tgDz == tianGanDiZhi_years[i]:
                return i
    
    result = []
    # 年
    yTG = day.getYearGZ(True)
    gz_year = Gan[yTG.tg] + Zhi[yTG.dz]
    result.append(getGZIndex(gz_year))
    
    # 月
    mTG = day.getMonthGZ()
    gz_month = Gan[mTG.tg] + Zhi[mTG.dz]
    result.append(getGZIndex(gz_month))
    
    # 日
    dTG = day.getDayGZ()
    gz_day = Gan[dTG.tg] + Zhi[dTG.dz]
    result.append(getGZIndex(gz_day))
    
    # 生肖
    result.append(yTG.dz)
    
    return result


def outAreaCode(result,areas):
    provinces = [
        "河北", "山西", "吉林", "辽宁", "黑龙江",
        "陕西", "甘肃", "青海", "山东", "福建",
        "浙江", "台湾", "河南", "湖北", "湖南",
        "江西", "江苏", "安徽", "广东", "海南",
        "四川", "贵州", "云南",
        "内蒙古", "广西", "西藏", "宁夏", "新疆",
        "北京", "天津", "上海", "重庆",
        "香港", "澳门"
    ]

    def outNumber(province, areas):
        if len(areas) == 0:
            return 0
        for i in range(len(areas) - 1):
            if province in areas[i]:
                return int(areas[i].replace(province, '').replace('注', ''))
        return 0
                
    for province in provinces:
        result.append(outNumber(province,areas))
    return result

x_data = []
y_data = []
for info in data['result']:
    date = info['date'].replace('(', ' ').replace(')', '').split(' ')[0].split('-')
    x_result = outTianGanDiZhi(int(date[0]), int(date[1]), int(date[2]))
    # x_result = outAreaCode(x_result,info['content'].split(','))
    # x_result.append(week_to_number(info['week']))
    x_data.append(x_result)
    
    y_result = list(map(int, info['red'].split(',')))
    y_result.append(int(info['blue']))
    y_data.append(y_result)

all_data = []
ball_0 = {'x_data': x_data, 'y_data': []}
for info in y_data:
    ball_0['y_data'].append(info[0])

ball_1 = {'x_data': [], 'y_data': []}
for info in y_data:
    ball_1['y_data'].append(info[1])
for i in range(len(x_data)):
    info = x_data[i].copy()
    info.append(y_data[i][0])
    ball_1['x_data'].append(info)
    
ball_2 = {'x_data': [], 'y_data': []}
for info in y_data:
    ball_2['y_data'].append(info[2])
for i in range(len(x_data)):
    info = x_data[i].copy()
    info.append(y_data[i][0])
    info.append(y_data[i][1])
    ball_2['x_data'].append(info)
    
ball_3 = {'x_data': [], 'y_data': []}
for info in y_data:
    ball_3['y_data'].append(info[3])
for i in range(len(x_data)):
    info = x_data[i].copy()
    info.append(y_data[i][0])
    info.append(y_data[i][1])
    info.append(y_data[i][2])
    ball_3['x_data'].append(info)

ball_4 = {'x_data': [], 'y_data': []}
for info in y_data:
    ball_4['y_data'].append(info[4])
for i in range(len(x_data)):
    info = x_data[i].copy()
    info.append(y_data[i][0])
    info.append(y_data[i][1])
    info.append(y_data[i][2])
    info.append(y_data[i][3])
    ball_4['x_data'].append(info)
    
ball_5 = {'x_data': [], 'y_data': []}
for info in y_data:
    ball_5['y_data'].append(info[5])
for i in range(len(x_data)):
    info = x_data[i].copy()
    info.append(y_data[i][0])
    info.append(y_data[i][1])
    info.append(y_data[i][2])
    info.append(y_data[i][3])
    info.append(y_data[i][4])
    ball_5['x_data'].append(info)
    
ball_6 = {'x_data': [], 'y_data': []}
for info in y_data:
    ball_6['y_data'].append(info[6])
for i in range(len(x_data)):
    info = x_data[i].copy()
    info.append(y_data[i][0])
    info.append(y_data[i][1])
    info.append(y_data[i][2])
    info.append(y_data[i][3])
    info.append(y_data[i][4])
    info.append(y_data[i][5])
    ball_6['x_data'].append(info)
final_data = {
    'ball_0': ball_0,
    'ball_1': ball_1,
    'ball_2': ball_2,
    'ball_3': ball_3,
    'ball_4': ball_4,
    'ball_5': ball_5,
    'ball_6': ball_6,
}

json_str = json.dumps(final_data, indent=4)
with open("data/svm-ssq/final_data.json", "w", encoding="utf-8") as f:
    f.write(json_str)
```

```python
import json
with open("data/svm-ssq/final_data.json", "r", encoding="utf-8") as f:
    train_data = json.load(f)
```
